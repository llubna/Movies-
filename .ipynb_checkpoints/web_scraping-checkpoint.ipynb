```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.boxofficemojo.com/chart/mpaa_title_lifetime_gross/?area=XWW&by_mpaa=G' 

response = requests.get(url)
```

```python
response.status_code 
```

```python
response.text[:1000]
```

```python
page = response.text
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
headers=[]
```

```python
for table in soup.find_all('div',class_='a-section imdb-scroll-table-inner'):
    for i in table.find_all('table',class_="a-bordered a-horizontal-stripes a-size-base a-span12 mojo-body-table mojo-table-annotated"):
        print(i.prettify())
    
```

```python
for table in soup.find_all('div',class_='a-section imdb-scroll-table-inner'):
    for element in table.find_all('table',class_="a-bordered a-horizontal-stripes a-size-base a-span12 mojo-body-table mojo-table-annotated"):
        for data in element.find_all('th'):
            headers.append(data.find('span').text)
            
            
```

```python
headers
```

```python
rows= [row for row in table.find_all('tr')]
```

```python
rows[1]
```

```python
rows[1].find_all('td')[0].find('a')
```

```python
movies = {}


for row in rows [1:1000]:
        items= row.find_all('td')
        link= items[0].find('a')
        title, url= link.text, link['href']
        movies[title] = [url] + [i.text for i in items]
        
movies
```
